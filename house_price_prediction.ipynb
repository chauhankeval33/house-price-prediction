Primary metric: RMSE, secondary: MAE and R²

Data Preprocessing

```python
#seprate feature and target
x = df.drop(columns=[target_col])
y = df[target_col]
```

```python
x.head()
```

```python
y.head()
```

```python
#train test spilt
x_train,x_test,y_train,y_test=train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=RANDOM_STATE
)
```

```python
print("train shape",x_train.shape)
print("test shape",x_test.shape)
```

Preprocessing Pipeline

```python
numerical_feature = x_train.select_dtypes(include=[np.number]).columns.tolist()
categorical_feature = x_train.select_dtypes(exclude=[np.number]).columns.tolist()

print("numerical feature",numerical_feature)
print("categorical feature",categorical_feature)

numeric_transformer = Pipeline([
   ( "imputer",SimpleImputer(strategy='median')),
   ("Scaler",StandardScaler())
])

categorical_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("scaler",OneHotEncoder(handle_unknown="ignore"))
])

# preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ("num",numeric_transformer,numerical_feature),
        ("col",categorical_transformer,categorical_feature)
    ]
)
```

```python
basline_pipe=Pipeline([
    ("preprocessor", preprocessor),
    ("model",LinearRegression())
])
```

```python
# preprocess the data and train the baseline model
basline_pipe.fit(x_train,y_train)
```

Evaluation for baseline model

```python
trainbaseline_pred=basline_pipe.predict(x_train)
testbaseline_pred=basline_pipe.predict(x_test)
```

```python
trainbaseline_pred[:5]
```

```python
y_train[:5]
```

```python
trainbaseline_rms = root_mean_squared_error(y_train,trainbaseline_pred)
trainbaseline_mae = mean_absolute_error(y_train,trainbaseline_pred)
trainbaseline_r2 = r2_score(y_train,trainbaseline_pred)

print("\n=== TRAIN BASELINE METRICS (LinearRegression) ===")
print(f"RMSE: {trainbaseline_rms:.3f}")
print(f"MAE : {trainbaseline_mae:.3f}")
print(f"R2  : {trainbaseline_r2:.3f}")
```

```python
testbaseline_rms = root_mean_squared_error(y_test,testbaseline_pred)
testbaseline_mae = mean_absolute_error(y_test,testbaseline_pred)
testbaseline_r2 = r2_score(y_test,testbaseline_pred)

print("\n=== TRAIN BASELINE METRICS (LinearRegression) ===")
print(f"RMSE: {testbaseline_rms:.3f}")
print(f"MAE : {testbaseline_mae:.3f}")
print(f"R2  : {testbaseline_r2:.3f}")
```

Model selection & Optimization

```python
models={
    "linearRegression":LinearRegression(),
    "Ridge":Ridge(random_state=RANDOM_STATE),
    "Lasso":Lasso(random_state=RANDOM_STATE,max_iter=10000),
    "RandomForest":RandomForestRegressor(),
    "HistGB":HistGradientBoostingRegressor()
}
```

```python
k=5
cv = KFold(n_splits=k,random_state=RANDOM_STATE,shuffle=True)
```

```python
scoring = {
    "rmse": "neg_root_mean_squared_error",
    "mae": "neg_mean_absolute_error",
    "r2": "r2"
}
```

```python
rows=[]

for name, model in models.items():
    pipe = Pipeline([
        ("preprocess",preprocessor),
        ("model",model)
    ])

    scores = cross_validate(pipe,x_train,y_train,cv=cv,scoring=scoring,n_jobs=1)

    rows.append({
        "model": name,
        "cv_rmse": -scores["test_rmse"].mean(),
        "cv_mae": -scores["test_mae"].mean(),
        "cv_r2": scores["test_r2"].mean()        
    })

# sort based on lowest rmse value
cv_results = pd.DataFrame(rows).sort_values("cv_rmse")
print("=== CV Model Comparison ===")
print(cv_results)
```

```python
cv_results
```

```python
best_row = cv_results.iloc[0]
best_model_name = best_row["model"]
best_rmse = best_row["cv_rmse"]

print("Best model based on CV RMSE:")
print("Model :", best_model_name)
print("CV RMSE:", best_rmse)
```

BEST MODEL: HistGradientBoostingRegressor

```python

hgb_pipe = Pipeline(
    [
        ("preprocessor",preprocessor),
        ("model",HistGradientBoostingRegressor(random_state=RANDOM_STATE))
    ]
)
```

```python
# hyperparameters combination
param_grid = {
    "model__learning_rate": [0.03, 0.05, 0.1],
    "model__max_depth": [None, 3, 6],
    "model__max_leaf_nodes": [15, 31, 63],
    "model__min_samples_leaf": [20, 50, 100],
    "model__l2_regularization": [0.0, 0.1, 1.0]
}
```

```python
grid = GridSearchCV(
    estimator=hgb_pipe,
    param_grid=param_grid,
    cv=cv,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1,
    verbose=1,
)
```

```python
grid.fit(x_train, y_train)
```

```python
print("\n=== TUNED HistGB (CV) ===")
print("Best CV RMSE:", -grid.best_score_)
print("Best params:", grid.best_params_)
```

Retraining with best params

```python
hgb_best = Pipeline([
    ("preprocessor",preprocessor),
    ("model",HistGradientBoostingRegressor(
        l2_regularization=0.1,
        learning_rate=0.1,
        max_depth=None,
        max_leaf_nodes=63,
        min_samples_leaf=20
    ))

])
```

```python
# train best model on entire training data (can also be done with refit=True in grid search)
hgb_best.fit(x_train,y_train)
```

Final Evaluation

```python
train_final_prediction = hgb_best.predict(x_train)

train_final_rmse = root_mean_squared_error(y_train,train_final_prediction)
train_final_mae = mean_absolute_error(y_train,train_final_prediction)
train_final_r2 = r2_score(y_train,train_final_prediction)

print("\n=== FINAL MODEL (Tuned HGB) Train Performance ===")
print(f"rmse:{train_final_rmse:.3f}")
print(f"mea:{train_final_mae:.3f}")
print(f"r2:{train_final_r2:.3f}")
```

```python
test_final_prediction = hgb_best.predict(x_test)

test_final_rmse = root_mean_squared_error(y_test,test_final_prediction)
test_final_mea = mean_absolute_error(y_test,test_final_prediction)
test_final_r2 = r2_score(y_test,test_final_prediction)

print("\n=== FINAL MODEL (Tuned HGB) test Performance ===")
print(f"rmse:{test_final_rmse:.3f}")
print(f"mea:{test_final_mea:.3f}")
print(f"r2:{test_final_r2:.3f}")
```

```python
# residual plot

residuals = y_test- test_final_prediction

plt.Figure(figsize=(6,4))
plt.scatter(test_final_prediction,residuals,s=10)
plt.axhline(0)
plt.title("Residuals vs Predictions")
plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.show()

plt.figure(figsize=(6, 4))
sns.histplot(residuals, bins=40, kde=True)
plt.title("Residual Distribution")
plt.xlabel("Residual")
plt.ylabel("Count")
plt.show()
```

Building a predictive system

```python
def predict_house_price(
    model,
    longitude: float,
    latitude: float,
    housing_median_age: float,
    total_rooms: float,
    total_bedrooms: float,
    population: float,
    households: float,
    median_income: float,
    ocean_proximity: str
) -> float:
    """
    Predict median_house_value for one new house.
    total_bedrooms can be np.nan (pipeline will impute).
    """
    new_row = pd.DataFrame([{
        "longitude": longitude,
        "latitude": latitude,
        "housing_median_age": housing_median_age,
        "total_rooms": total_rooms,
        "total_bedrooms": total_bedrooms,
        "population": population,
        "households": households,
        "median_income": median_income,
        "ocean_proximity": ocean_proximity
    }])

    return float(model.predict(new_row)[0])
```

```python
# Example inference
example_pred = predict_house_price(
    model=hgb_best,
    longitude=-122.230,
    latitude=37.880,
    housing_median_age=41,
    total_rooms=880,
    total_bedrooms=129,
    population=322,
    households=126,
    median_income=8.3252,
    ocean_proximity="NEAR BAY"
 )

print("\nExample prediction:", round(example_pred, 2))
```
